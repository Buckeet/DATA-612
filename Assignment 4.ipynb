```python
#imported pandas, matplot library, and seaborn
#imported the dataset for baseballing batting statistics
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#for this assignment, I'm going to use batting data
#specifically, hits and strike outs
batting = pd.read_csv\
('https://raw.githubusercontent.com/frankData612/data_612/master/baseballdatabank-master/core/Batting.csv')

#And checked the data to make sure this is the data I what i want to use.
print(batting)
```

```python
# I want to look at the just the 2019 season
#so i reduced the dataset to only rows with a yearID of 2019
b1 = batting[batting['yearID'] == 2019]

#Then, I want to remove hitters that have little to no at-bats
#there is no official minimum number of at-bats for baseball statistics
#However, the MLB sets a minimum of 130 at-bats to be eligible for a rookie batting title
#So, I'm going to use 130 or more at bats for this data
b2019 = b1[b1['AB'] > 129]
print(b2019)
```

```python
#Here, I used the matplot library to show
#a scatterplot of the bivariate data: hits and strike outs
sp1 = plt.figure()
h_so = sp1.add_subplot()
h_so.scatter(b2019['H'], b2019['SO'])
h_so.set_title('Hits Versus Strike Outs')
h_so.set_xlabel('Hits')
h_so.set_ylabel('Strike Outs')
plt.show()
```

```python
#Here, I used the matplot library to create a histogram
#showing the frequency of hits in the dataset
hist1 = plt.figure()
hits = hist1.add_subplot()
hits.hist(b2019['H'], bins=18)
hits.set_title('Hits')
hits.set_xlabel('Number of Hits in 2019')
hits.set_ylabel('Frequency')
plt.show()
```

```python
#Here, I used the matplot library to create a histrogram 
#showing the frequency of Strike Outs in the dataset
hist2 = plt.figure()
SO = hist2.add_subplot()
SO.hist(b2019['SO'], bins=18)
SO.set_title('Strike Outs')
SO.set_xlabel('Number of Strike Outs in 2019')
SO.set_ylabel('Frequency')
plt.show()
```

```python
#Here, I used the matplots library to create a boxplot
#showing data on hits and strike outs
bp1 = plt.figure()
h_so1=bp1.add_subplot()
h_so1.boxplot([b2019['H'], b2019['SO']], labels=['Hits', 'Strike Outs'])
h_so1.set_title("Boxplot of Hits and Strike Outs")
plt.show()
```

```python
#Now, I want to show data using the seaborn library
#first, I make sure pandas, matplot library, and seaborn are imported
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Using the seaborn .distplot fuction, I create a density curve and histogram for the univariate data 'Hits' in our dataset
hist, ax = plt.subplots()
ax = sns.distplot(b2019['H'])
ax.set_title('Hits Per Player Histogram and Density Plot')
ax.set_xlabel('Hits')
plt.show()

# this type of chart is best for showing univariate date
```

```python
#Here is the same type of chart but for 'Strike Out' Data
hist, ax = plt.subplots()
ax = sns.distplot(b2019['SO'])
ax.set_title('Strike Outs Per Player Histogram and Density Plot')
ax.set_xlabel('Strike Outs')
plt.show()
```

```python
#Here, I use the seaborn libary to crete a regression plot using the .regplot function
#I plotted Hits on the x-axis and Strike Outs on the y-axis
#scatterplot/regressionplot is best for this data because the data is bivariate; hits and strike outs

scatter, ax = plt.subplots()
ax = sns.regplot(x=b2019['H'], y=b2019['SO'])
ax.set_title('Hits Versus Strike Outs')
ax.set_xlabel('Hits')
ax.set_ylabel('Strike Outs')
plt.show()

#As you can see from the the regression plot, this data has a positive linear correlation
```

```python
#Since the data we are using is very dense, we can use a hexbinplot to compare the bivariate data
#I used the .jointplot fuction
#I set hits to the x-axis and strike outs to the y-axis

hexbin = sns.jointplot(x=b2019['H'], y=b2019['SO'], kind="hex")
hexbin.set_axis_labels(xlabel='Hits', ylabel='Strike Outs')
hexbin.fig.suptitle('Hexbin Joint Plot of Hits and Strike Outs', fontsize=12, y=1.02)
plt.show()
```

```python
#Another way to show Hits and Strike Outs is using a Kernal Density Plot
#using seaborn's .kdeplot function
#again, like the Hexbin Plot, this is a good way to show bivariate data that is dense

kde, hso = plt.subplots()
hso = sns.kdeplot(x=b2019['H'], y=b2019['SO'], shade=True)
hso.set_title('Kernal Density Plot of Hits and Strike Outs 2019')
hso.set_xlabel('Hits')
hso.set_ylabel('Strike Outs')

#And here is the same kernal density plot with density curves
#using seaborn's .jointplot function, rather than .kdeplot

kde_hso = sns.jointplot(x=b2019['H'], y=b2019['SO'], kind='kde', shade=True)
kde_hso.set_axis_labels(xlabel='Hits', ylabel='Strike Outs')
kde_hso.fig.suptitle('Kernal Density Plot of Hits and Strike Outs 2019 with Density Curves', y=1.02)
plt.show()
```

```python
#Now, going back to univariate data
#I want to create a boxplot and a violinplot for Hits data
#For the boxplot I used the .boxplot function in the Seaborn library
bar, h = plt.subplots()
h = sns.boxplot(y=b2019['H'])
h.set_title('Boxplot of Hits by Player in 2019')
h.set_xlabel('Hits')
h.set_ylabel('Number of Hits')
plt.show()

#For the violinplot, I used the .violinplot function in the Seaborn library
h = sns.violinplot(y=b2019['H'])
h.set_title('Violinplot of Hits by Player in 2019')
h.set_xlabel('Hits')
h.set_ylabel('Number of Hits')
plt.show()
```

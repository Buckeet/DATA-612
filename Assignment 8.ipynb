```python
# import pandas
import pandas as pd

# import baseball manager csv file
mgt = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/baseballdatabank-master/core/Managers.csv')
mgt.head()
```

```python
# create summary data function to create summary dataframe for groupby data
# looking at total wins, total losses, and total games
# Win %, first and last year as manager, total years as manager, and average rank

def summarize_data(g):
    return(
        pd.Series({
            "Total_Wins": g.W.sum(),
            "Total_Losses": g.L.sum(),
            "Total_Games": g.G.sum(),
            "Win_Percentage": g.W.sum()/g.G.sum(),
            "First_Yr_Mgr": g.yearID.min(),
            "Last_Yr_Mgr": g.yearID.max(),
            "Yrs_As_Mgr": len(g),
            "Avg_Rank": g['rank'].mean(),
    })
  )

mgt_summary = mgt.groupby("playerID").apply(summarize_data)
mgt_summary
```

```python
mgt_summary.to_csv('mgt_summary.csv')
```
